```python
# 픽률 그래프 그리기 (평균치 있음@@)
# 반복문으로 코드 압축
Tier = ["All", "Bronze", "Silver", "Gold", "Platinum", "Diamond", "Master", "Grandmaster"]
heroes = ['Ana', 'Mercy', 'Kiriko', 'Moira', 'Zenyatta', 'Lucio', 'Baptiste', 'Brigitte']

plt.figure(figsize=(8,8))
for hero in heroes:
    ow2_sup_fisrt_hero = ow2_sup_fisrt[ow2_sup_fisrt["Hero"] == hero]
    plt.plot(ow2_sup_fisrt_hero["Skill Tier"], ow2_sup_fisrt_hero["Pick Rate, %"], label=hero)

#평균치 그리기
no_sort_mean = ow2_sup_fisrt["Pick Rate, %"].groupby(ow2_sup_fisrt["Skill Tier"]).mean()
sort_mean_PickRate = no_sort_mean.reindex(Tier)
plt.plot(sort_mean_PickRate, linestyle='dashed', linewidth=2, color='black', marker='o', markersize=3 )

plt.xlabel("플레이어 티어")
plt.ylabel("픽률(%)")
plt.title("티어에 따른 서포터 영웅의 픽률 분포")
plt.legend(['Ana', 'Mercy', 'Kiriko', 'Moira', 'Zenyatta', 'Lucio', 'Baptiste', 'Brigitte', 'mean'], loc=[1.02,0.53])

plt.show()
```
