```python
import matplotlib.pyplot as plt
from matplotlib.offsetbox import OffsetImage, AnnotationBbox
import numpy as np

# 이미지 파일 경로 ("./test_image.png")
image_path = "./test_image.png"

# 데이터 예시 (파이 그래프를 그릴 때 사용할 데이터)
data = [25, 35, 20, 10, 10]

# 파이 그래프 그리기
fig, ax = plt.subplots()
ax.pie(data, labels=['A', 'B', 'C', 'D', 'E'], autopct='%1.1f%%', startangle=90)

# 이미지를 파이 그래프 위에 추가
for i, val in enumerate(data):
    theta1 = np.deg2rad(sum(data[:i]))
    theta2 = np.deg2rad(sum(data[:i+1]))
    angle = (theta1 + theta2) / 2

    # 이미지를 OffsetImage로 변환
    img = plt.imread(image_path)
    imagebox = OffsetImage(img, zoom=0.05)

    # AnnotationBbox를 사용하여 이미지를 파이 조각 위에 추가
    ab = AnnotationBbox(imagebox, (0.5 + 0.4 * np.cos(angle), 0.5 + 0.4 * np.sin(angle)),
                        frameon=False, boxcoords="axes fraction", pad=0.1)
    ax.add_artist(ab)

plt.show()
```
