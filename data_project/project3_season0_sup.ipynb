```python
# Path of the file to read
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
```

```python
# 그래프 한글 설정
plt.rc("font", family = "Malgun Gothic")
sb.set(font="Malgun Gothic",
rc={"axes.unicode_minus":False}, style='white')
```

```python
# 사용될 영웅, 티어 리스트 
Tier = ["All", "Bronze", "Silver", "Gold", "Platinum", "Diamond", "Master", "Grandmaster"]
Sup_heroes = ['Ana', 'Mercy', 'Kiriko', 'Moira', 'Zenyatta', 'Lucio', 'Baptiste', 'Brigitte']
```

```python
# 게임 출시 이후 ~ 랭크 전 데이터
ow2_data_0_path = "../data_project/data/ow2_quickplay_heroes_stats.csv"
ow2_data_0 = pd.read_csv(ow2_data_0_path, index_col= "Role")
ow2_data_0
```

```python
#역할군이 서포터로 분류되어 있는 애들만 추출해오기
ow2_data_0_sup = ow2_data_0.loc[ow2_data_0.index == "Support"]
ow2_data_0_sup

#일반적인 지표 값만 가져오기 ~11, 컬럼12 이후로는 캐릭터 별 스킬에 대한 수치가 나타나있음. ex) 궁극기 사용 횟수
ow2_ss0_sup = ow2_data_0_sup[ow2_data_0_sup.columns[:11]]
ow2_ss0_sup
```

```python
# 픽률 그래프 그리기 (평균치 있음@@)
# 반복문으로 코드 압축
plt.figure(figsize=(8,8))
for hero in Sup_heroes:
    ow2_sup_fisrt_hero = ow2_ss0_sup[ow2_ss0_sup["Hero"] == hero]
    plt.plot(ow2_sup_fisrt_hero["Skill Tier"], ow2_sup_fisrt_hero["Pick Rate, %"], label=hero)

plt.xlabel("플레이어 티어")
plt.ylabel("픽률(%)")
plt.title("티어에 따른 서포터 영웅의 픽률 분포")
plt.legend(Sup_heroes, loc=[1.02,0.53])

plt.show()
```

```python
# 픽률 그래프 그리기 (평균치 있음@@)
# 반복문으로 코드 압축
plt.figure(figsize=(8,8))
for hero in Sup_heroes:
    ow2_sup_fisrt_hero = ow2_ss0_sup[ow2_ss0_sup["Hero"] == hero]
    plt.plot(ow2_sup_fisrt_hero["Skill Tier"], ow2_sup_fisrt_hero["Pick Rate, %"], label=hero)

#평균치 그리기
no_sort_mean = ow2_ss0_sup["Pick Rate, %"].groupby(ow2_ss0_sup["Skill Tier"]).mean()
sort_mean_PickRate = no_sort_mean.reindex(Tier)
plt.plot(sort_mean_PickRate, linestyle='dashed', linewidth=2, color='black', marker='o', markersize=3 )

plt.xlabel("플레이어 티어")
plt.ylabel("픽률(%)")
plt.title("티어에 따른 서포터 영웅의 픽률 분포")
plt.legend(['Ana', 'Mercy', 'Kiriko', 'Moira', 'Zenyatta', 'Lucio', 'Baptiste', 'Brigitte', 'mean'], loc=[1.02,0.53])

plt.show()
```

```python
#승률 그래프 그리기
# 반복문으로 코드 압축
plt.figure(figsize=(8,8))
for hero in Sup_heroes:
    ow2_sup_fisrt_hero = ow2_ss0_sup[ow2_ss0_sup["Hero"] == hero]
    plt.plot(ow2_sup_fisrt_hero["Skill Tier"],
              ow2_sup_fisrt_hero["Win Rate, %"],
              label=hero)


plt.xlabel("플레이어 티어")
plt.ylabel("승률 (%)")
plt.title("티어에 따른 서포터 영웅의 승률 분포")
plt.legend(Sup_heroes, loc=[1.02,0.53])
plt.show()
```

```python
#힐량 평균 구하기
heal_mean = ow2_ss0_sup["Healing / 10min"].groupby(ow2_ss0_sup["Hero"]).mean()

# 그래프 그리기
plt.figure(figsize=(8, 8))  
sb.barplot(x =Sup_heroes, y=heal_mean)
plt.ylim(5000, 9000)  #가시성을 위해 5천이하의 값 삭제
plt.xlabel("영웅")
plt.ylabel("힐량 / 10분")
plt.title("서포터 캐릭별 10분당 힐량")
```
