```python
def Draw_PickRate():
    user_input =input("알고싶은 영웅의 이름")
    User_graph = pd.DataFrame(columns=['season', 'Pick Rate, %'])
    i = 0
    # 순환문을 통해 각 파일의 데이터 추출
    for file_address in season_data_path_list:
        # 파일 읽기
        data = pd.read_csv(file_address)

        # 원하는 컬럼의 값만 선택
        selected_column = ['Hero', 'Pick Rate, %']
        selected_values = data[selected_column]

        # 해당 시즌의 영웅별 픽률 평균 구하기
        season_pick_avg = selected_values["Pick Rate, %"].groupby(data["Hero"]).mean()

        # 그 픽률 평균 중에 특정 영웅의 값만 추출
        select_hero_pickrate = season_pick_avg[user_input] # "Ana" 일 때 7.16875

        #print(file_address, "은" , select_hero)

        # 꿀 수치 기반 탱커 데이터프레임
        result = str(i) + "시즌"
        insert_data = [ result , select_hero_pickrate ]
        User_graph.loc[len(User_graph)] = insert_data
        i += 1

    #User_graph

    # 그래프 그리기
    plt.figure(figsize=(8, 8))  
    plt.plot(User_graph['season'], User_graph['Pick Rate, %'], marker='.', label='픽률')
    plt.grid(axis='x') # y축 그리드뜨뜨뜨뜨
    plt.ylabel("픽률")
    title_input =  user_input + " 의 시즌 별 픽률"
    plt.title(title_input)

    return 0

Draw_PickRate()
```

```python
def Draw_Rate():
    user_input =input("알고싶은 영웅의 이름")
    User_graph = pd.DataFrame(columns=['season', 'Pick Rate, %', 'Win Rate, %'])
    i = 0
    # 순환문을 통해 각 파일의 데이터 추출
    for file_address in season_data_path_list:
        # 파일 읽기
        data = pd.read_csv(file_address)

        # 원하는 컬럼의 값만 선택
        selected_column = ['Hero', 'Pick Rate, %', 'Win Rate, %']
        selected_values = data[selected_column]

        # 해당 시즌의 영웅별 픽률 평균 구하기
        season_pick_avg = selected_values["Pick Rate, %"].groupby(data["Hero"]).mean()
        season_win_avg = selected_values["Win Rate, %"].groupby(data["Hero"]).mean()

        # 그 픽률 평균 중에 특정 영웅의 값만 추출
        select_hero_pickrate = season_pick_avg[user_input] # "An" 일 때 7.16875
        select_hero_winrate = season_win_avg[user_input]

        # 꿀 수치 기반 탱커 데이터프레임
        result = str(i) + "시즌"
        insert_data = [ result ,select_hero_pickrate, select_hero_winrate]
        User_graph.loc[len(User_graph)] = insert_data
        i += 1

    # 가충치 부여해서 그래프 격차 줄이기.
    test = User_graph['Win Rate, %'].max() - User_graph['Pick Rate, %'].min()
    #print(test)

    # 그래프 그리기
    plt.figure(figsize=(8, 8))  
    plt.plot(User_graph['season'], User_graph['Pick Rate, %']+ test, marker='.', label='픽률')
    plt.plot(User_graph['season'], User_graph['Win Rate, %'], marker='.', label='승률')
    plt.grid(axis='x') # y축 그리드뜨뜨뜨뜨
    plt.ylabel("승률")
    plt.legend()
    title_input =  user_input + " 의 시즌 별 그래프, 픽률 가중치 " + str(test) #str(40) #str(test)
    plt.title(title_input)

    return 0

Draw_Rate()
```
