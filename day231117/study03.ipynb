```python
import pandas as pd
import folium

data_csv = pd.read_csv("../택시승차대 현황.csv", encoding='cp949')

q2_data = pd.DataFrame(data_csv)

q2 = folium.Map(location=[37.547631, 126.942463], zoom_start=15)

for index, row in q2_data.iterrows():
    folium.Marker(location= [row["위도"], row["경도"]] , 
                  tooltip=row["위치명"],
                  icon=folium.Icon(color='red', icon='fa-solid fa-taxi', prefix='fa-solid')
                  ).add_to(q2)
    folium.Circle([row["위도"], row["경도"]],
                   color='red',
                   radius=50
                   ).add_to(q2)

q2
```

```python
import pandas as pd
import folium

data_csv = pd.read_csv("../택시승차대 현황.csv", encoding='cp949')

print( data_csv.head())
taxi = data_csv.loc[:, ["위치명", "위도", "경도"]]

q2 = folium.Map(location=[37.547631, 126.942463], zoom_start=15, width=500, height=500)


for index, row in taxi.iterrows():    
    folium.Marker([row["위도"], row["경도"]],
                   tooltip=row["위치명"], 
                   ).add_to(q2)
    folium.Circle([row["위도"], row["경도"]],
                   color='red',
                   radius=50
                   ).add_to(q2)

q2
```
