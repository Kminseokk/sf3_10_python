```python
import numpy as np

x = np.arange(3)

plt.bar(x, [100,200,300], color=['r','g','b'], width=0.6)
plt.xticks(x, ['2016', '2017','2018'])
plt.show()
```

```python
n = 100
x = np.random.rand(n)
y = np.random.rand(n)

area = (30 * np.random.rand(n)) ** 2
colors = np.random.rand(n)

plt.scatter(x, y, s=area, c=colors, cmap="Spectral", alpha=0.5)
plt.show()
```

```python
ratio = [34,32,16,18]
labels = ['Apple', 'Banana', 'Melon', 'Grapes']
wedgeprops = { 'width' : 0.7, 'edgecolor':'k', 'linewidth':1}

plt.pie(ratio, labels=labels, autopct='%.1f%%', explode=[0, 0.1, 0, 0.1], colors=['red', 'yellow', 'green', 'purple'], wedgeprops=wedgeprops)
```

```python
import matplotlib.pyplot as plt

months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
sales_2019 = [100,120,140,110,130,150,160,170,180,200,190,210]
sales_2020 = [90,110,130,120,140,160,170,160,150,180,200,190]

plt.plot(months, sales_2019, label='2019')
plt.plot(months, sales_2020, label='2020')
plt.xlabel('Month', loc='center')
plt.ylabel('Sales', loc='center')


plt.legend()
plt.show()
```

```python
categories = ['Category 1','Category 2','Category 3','Category 4','Category 5']
data = [20, 35, 15, 27, 45]

x = np.arange(5)

plt.title('bar chart')
plt.bar(x, data)
plt.xticks(x, categories, rotation=45)
plt.ylim(0,50)
plt.xlabel('categories', loc='center')
plt.ylabel('values', loc='center')
plt.grid(True)
plt.show()
```

```python

x = np.arange(0,10,0.01)
y = np.sin(x)

plt.plot(x, np.sin(x), label='sin(x)')
plt.plot(x, np.cos(x), label='cos(x)')
plt.xlabel('x', loc='center')
plt.ylabel('y', loc='center')
plt.grid(True)

plt.legend()
plt.show()
```
